```python
import numpy as np
from SIRD import SIRD
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
def plot(I0,beta,gamma,delta1,delta2,K,severe,N):
    t_eval = np.arange(365)
    y = SIRD(t_eval,I0,beta,gamma,delta1,delta2,K,severe)
    fig = plt.figure()
    labels = ['Susceptible','Infected','Recovered','Died']
    for i in range(4):
        plt.plot(t_eval,y[i],label=labels[i])
    plt.plot(t_eval,[10*K]*len(t_eval),'--k',label='Healthcare Capacity')
    plt.ylim(0,1)
    plt.legend()
    plt.title(f"$R_0 = {beta/gamma:.1f}$, Peak Infected: {100*y[1].max():.1f}%, Death Rate: {100*y[3,-1]/y[2,-1]:1.1f}%")
    plt.show()
```

```python
plot(.001,1.4/14,1/14,.01,.08,.005,.1,330000000)
```
